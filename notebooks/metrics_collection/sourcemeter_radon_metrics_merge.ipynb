```python
import os
os.mkdir("/content/Methods_Metrics_Part3/", mode=0o777)
```

```python
TXT_FILENAME = 'em_repo_stars_part3.txt'
with open(TXT_FILENAME) as f:
    repos = [line.strip() for line in f.readlines()]
```

```python
import pandas as pd
from tqdm import tqdm

for repo in tqdm(repos):
  print(repo)
  repo_file_name = repo.removeprefix("https://github.com/").replace("/", "_")
  sm_methods = pd.read_csv(f'./SM_Part3_Method/{repo_file_name}-Method.csv')
  radon_methods = pd.read_csv(f'Radon_Part3_Methods/{repo_file_name}-Methods.csv')

  fullpaths = []
  for id in sm_methods.index:
    fullpath = sm_methods["Path"][id].removeprefix("/content/")+"~"+str(sm_methods["Name"][id])
    fullpaths.append(fullpath)
  sm_methods["Fullpath"] = fullpaths

  fullpaths = []
  for id in radon_methods.index:
    fullpath = radon_methods["path"][id]+"~"+str(radon_methods["name"][id])
    fullpaths.append(fullpath)
  radon_methods["Fullpath"] = fullpaths

  joint_metrics = pd.merge(sm_methods, radon_methods,
                    on='Fullpath',
                    how='inner')
  joint_metrics = joint_metrics.drop(["name", "path", "Fullpath"], axis=1)

  joint_metrics.to_csv(f"/content/Methods_Metrics_Part3/{repo_file_name}_metrics.csv")

  # displaying result
  print(joint_metrics)
```
