```python
import pandas as pd
import re
```

```python
k_feature_reductions = []
current_k = None

with open("outputs/log/abc_meta_90.log", "r") as f:
    for line in f:
        k_match = re.search(r"Reduction to (\d+) features", line)
        if k_match:
            current_k = int(k_match.group(1))
        subset_match = re.search(r"Selected subset: \[(.*?)\]", line)
        if subset_match and current_k is not None:
            subset = tuple(sorted(subset_match.group(1).replace("'", "").split(", ")))
            k_feature_reductions.append((current_k, subset))

df_reductions = pd.DataFrame(k_feature_reductions, columns=["k", "Subset"])

df_subset_counts = df_reductions.groupby(["k", "Subset"]).size().reset_index(name="Frequency")
df_total_per_k = df_reductions.groupby("k").size().reset_index(name="Total")
df_subset_counts = pd.merge(df_subset_counts, df_total_per_k, on="k")
df_subset_counts["Proportion"] = df_subset_counts["Frequency"] / df_subset_counts["Total"]

df_subset_counts = df_subset_counts.sort_values(by=["k", "Proportion"], ascending=[True, False])

df_repeated_subsets = df_subset_counts[df_subset_counts["Proportion"] > 0.1]
df_repeated_subsets_sorted = df_repeated_subsets.sort_values(by=["Proportion", "k"], ascending=[False, True])

df_repeated_subsets_sorted.head(15)
```
