```python
import pandas as pd
import re
```

```python
k_feature_reductions = []
current_k = None
best_error = None

with open("../../outputs/log/abc_total_corr_90.log", "r") as f:
    for line in f:
        k_match = re.search(r"Reduction to (\d+) features", line)
        if k_match:
            current_k = int(k_match.group(1))
        
        error_match = re.search(r"Best error: ([0-9\.eE\-]+)\.", line)
        if error_match:
            best_error = float(error_match.group(1))
        
        subset_match = re.search(r"Selected subset: \[(.*?)\]", line)
        if subset_match and current_k is not None and best_error is not None:
            subset = tuple(sorted(subset_match.group(1).replace("'", "").split(", ")))
            k_feature_reductions.append((current_k, subset, best_error))

df_reductions = pd.DataFrame(k_feature_reductions, columns=["k", "Subset", "Best Error"])
df_subset_counts = df_reductions.groupby(["k", "Subset"]).size().reset_index(name="Numerator")
df_total_per_k = df_reductions.groupby("k").size().reset_index(name="Denominator")
df_subset_counts = pd.merge(df_subset_counts, df_total_per_k, on="k")
df_subset_counts["Frequency"] = df_subset_counts["Numerator"] / df_subset_counts["Denominator"]
df_subset_counts = df_subset_counts.sort_values(by=["Frequency", "k"], ascending=[False, True])

df_final = pd.merge(
    df_subset_counts,
    df_reductions[["k", "Subset", "Best Error"]],
    on=["k", "Subset"],
    how="left"
)

df_final_mean = df_final.groupby("Subset", as_index=False).agg({
    "Best Error": "mean",
    "Frequency": "mean",
    "k": "mean"
})

df_final_mean = df_final_mean.sort_values(by=["Frequency", "k"], ascending=[False, True])

df_final_mean.head(15)
```
